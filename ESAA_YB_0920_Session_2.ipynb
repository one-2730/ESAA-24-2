### Step 1. Import the necessary libraries

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

### Step 2. Create the DataFrame that should look like the one below.
*참고*

```
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}
```

````python
score_df = pd.DataFrame({'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]})

score_df
````

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

````python
plt.figure(figsize = (5, 3))

sns.scatterplot(x = "preTestScore", y = "postTestScore", data = score_df, size = "age")

plt.title("Relationship btw preTestScore & postTestScore", fontsize = 10)
plt.xlabel("preTestScore", fontsize = 8)
plt.ylabel("postTestScore", fontsize = 8)
plt.show()
````

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

````python
plt.figure(figsize = (5, 3))

sns.scatterplot(x = "preTestScore", y = "postTestScore", data = score_df, size = 4.5*score_df["postTestScore"], hue = "female")

plt.title("Relationship btw preTestScore & postTestScore")
plt.show()
````
